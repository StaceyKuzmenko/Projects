### Объединение датафреймов в единый датафрейм

```python
final_orders = (
    orders.sort_values(by=['date'])
    .groupby('user_id')
    .agg(
        {
            'date': 'first',
            'order_id': 'count',
            'revenue': 'sum',
            'quantity': 'sum',
            }
        )
    .rename(columns={'order_id': 'order_number'})
    .reset_index()    
)


df_info(final_orders)
final_orders.describe()
```

```python
final_funnel = (
    ab_funnel.sort_values(by=['date'])
    .groupby('user_id')
    .agg(
        {
            'date': 'first',
            'session_start': 'sum',
            'view_item': 'sum',
            'add_to_cart': 'sum',
            'view_cart': 'sum', 
            'checkout': 'sum',
            }
        )
    .reset_index()    
)


df_info(final_funnel)
final_funnel.describe()
```

```python
final_data = users.merge(final_orders, on='user_id', how='left')
final_data.info()
```

```python
#Удалим ненужный столбец с датой
del final_data['date']
final_data.info()
```

```python
final_data = final_data.merge(final_funnel, on='user_id', how='left')
final_data.info()
```

```python
#У нас появилась одна строка с пропусками почти по всем параметрам. Посмотрим на нее и удалим (она не повлияет на общую картину).
display(final_data[final_data['checkout'].isna()])
```

```python
final_data = final_data[final_data['user_id'] != 'xh9rDFwo1X']
final_data.info()
```

```python
#Для удобства поменяем местами столбцы в датафрейме
final_data = (
    final_data[
        ['user_id', 'exp_group', 'date', 'session_start', 'view_item', 'add_to_cart', 'view_cart', 'checkout', 'order_number', 'revenue', 'quantity']
    ]
)
final_data
```

```python
#Напишем функцию замены нулей на NaN в столбцах, так как они влияют на расчет средних показателей (в отличие от NaN)
def replace_0_nan(cols):
    final_data[cols] = final_data[cols].mask(final_data[cols].eq(0))
```

```python
replace_0_nan('session_start')
replace_0_nan('view_item')
replace_0_nan('add_to_cart')
replace_0_nan('view_cart')
replace_0_nan('checkout')
```

```python
final_data_1 = final_data
```

```python
final_data_1 = (
    final_data_1
    .melt(id_vars=['user_id', 'exp_group', 'date'], value_vars=['session_start', 'view_item', 'add_to_cart', 'view_cart', 'checkout', 'order_number', 'revenue', 'quantity'], ignore_index=False)
)    
```

```python
final_data_for_funnel = final_data_1[final_data_1['variable'] != 'revenue']
final_data_for_funnel = final_data_for_funnel[final_data_for_funnel['variable'] != 'quantity']
```

```python
final_data_for_funnel.info()
```

## Часть. Анализ воронки событий

```python
funnel_df = final_data_for_funnel.groupby('variable')['value'].sum().reset_index().sort_values('value', ascending=False)
funnel_df = funnel_df.reset_index(drop=True)
funnel_df.iloc[0,:], funnel_df.iloc[1,:] = funnel_df.iloc[1,:], funnel_df.iloc[0,:]
funnel_df.iloc[2,:], funnel_df.iloc[3,:] = funnel_df.iloc[3,:], funnel_df.iloc[2,:]
display(funnel_df)
```

```python
# Строим график воронки
fig = go.Figure(go.Funnel(
        name = 'test',
        y = funnel_df['variable'],
        x = funnel_df['value'],
        textposition = "inside",
        textinfo = "value+percent previous+percent initial"))
fig.update_layout(
    title={
        'text': 'Воронка по событиям, совершенным всеми пользователями',
        'y':0.88,
        'x':0.53,
        'xanchor': 'center',
        'yanchor': 'top'},
    )
fig.show()
```

Если рассматривать воронку продаж по всем пользователям (без разделения на группы), то **общая конверсия составляет 6.6%.** Больше всего пользователей совершают событий на этапе просмотра карточек товаров (в 2,5 раза больше в сравнии с началом сессии). **Больше всего пользователей "отваливается" на моменте "чекаута"**: до покупки доходит меньше 20% из них. Значит, есть какая-то проблема с этапом чекаута. Возможно, от пользователей требуется полная регистрация на сайте, с большим количеством параметров, которые нужно заполнить. Клиентов это не устраивает и они покидают сайт.

## Предварительная проверка полученных данных A/B теста (A/A тест)

Критерии успешного A/A-теста:

- Количество пользователей в различных группах различается не более, чем на 1%;
- Для всех групп фиксируют и отправляют в системы аналитики данные об одном и том же;
- Различие ключевых метрик по группам не превышает 1% и не имеет статистической значимости;
- Попавший в одну из групп посетитель остаётся в этой группе до конца теста. Если пользователь видит разные версии сайта (и с новым форматом промо-акций, и без) в ходе одного исследования, неизвестно, какая именно повлияла на его решения. Значит, и результаты такого теста нельзя интерпретировать однозначно.

```python
count_visitors_a = group_a['user_id'].nunique()
count_visitors_b = group_b['user_id'].nunique()
groups_comparison = count_visitors_b / count_visitors_a - 1
print('Количество пользователей в группе A:', count_visitors_a)
print()
print('Количество пользователей в группе B:', count_visitors_b)
print()
print('Кол-во пользователей в группе B больше кол-ва пользователей группы A на:', f'{groups_comparison:.0%}')
```

Количество пользователей в разных группах **различается на 2%, что не является нормой для проведения теста.** Также, в самом начале исследования мы отметили, что более 50 пользователей пересекались в группах (мы их удалили). 

**Данные недочеты не допустимы при проведении A/B теста.** По-хорошему, необходимо прервать тестирование, так как его результаты будут некорректные. Но так как у нас нет возможности получить иные данные, продолжим оценивать имеющиеся.

## Статистическая значимость различий в конверсии и в выручке между группами

```python
#подготовим данные к проверке гипотез
visitors_a_daily = final_data[final_data['exp_group'] == 0][['date', 'session_start']]
visitors_a_daily = visitors_a_daily.groupby('date')['session_start'].sum().reset_index()
visitors_a_daily.columns = ['date', 'visitors_per_date_a']

visitors_a_cummulative = visitors_a_daily.apply(
    lambda x: visitors_a_daily[visitors_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'visitors_per_date_a': 'sum'}
    ),
    axis=1,
)
visitors_a_cummulative.columns = ['date', 'visitors_cummulative_a']

visitors_b_daily = final_data[final_data['exp_group'] == 1][['date', 'session_start']]
visitors_b_daily = visitors_b_daily.groupby('date')['session_start'].sum().reset_index()
visitors_b_daily.columns = ['date', 'visitors_per_date_b']

visitors_b_cummulative = visitors_b_daily.apply(
    lambda x: visitors_b_daily[visitors_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'visitors_per_date_b': 'sum'}
    ),
    axis=1,
)
visitors_b_cummulative.columns = ['date', 'visitors_cummulative_b']

orders_a_daily = (
    final_data[final_data['exp_group'] == 0][['date', 'order_number', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'order_number': pd.Series.nunique, 'revenue': 'sum'})
)
orders_a_daily.columns = [
    'date', 
    'orders_per_date_a', 
    'revenue_per_date_a'
]

orders_a_cummulative = orders_a_daily.apply(
    lambda x: orders_a_daily[orders_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_a': 'sum', 'revenue_per_date_a': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
orders_a_cummulative.columns = [
    'date',
    'orders_cummulative_a',
    'revenue_cummulative_a',
]

orders_b_daily = (
    final_data[final_data['exp_group'] == 1][['date', 'order_number', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'order_number': pd.Series.nunique, 'revenue': 'sum'})
)
orders_b_daily.columns = ['date', 'orders_per_date_b', 'revenue_per_date_b']

orders_b_cummulative = orders_b_daily.apply(
    lambda x: orders_b_daily[orders_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_b': 'sum', 'revenue_per_date_b': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
orders_b_cummulative.columns = [
    'date',
    'orders_cummulative_b',
    'revenue_cummulative_b',
]

data = (
    orders_a_daily.merge(
        orders_b_daily, left_on='date', right_on='date', how='left'
    )
    .merge(orders_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(orders_b_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_a_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_b_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_b_cummulative, left_on='date', right_on='date', how='left')
)

display(data.tail(5)) 
```

### Расчет статистической значимости различия конверсии между группами

Сформулируем гипотезы.

`Нулевая`: различий в конверсии между группами нет.

`Альтернативная`: различия в конверсии между группами есть.

```python
#Посчитаем статистическую значимость различия в конверсии между группами

#В переменных укажем число совершенных заказов пользователями, которые заказывали хотя бы 1 раз
orders_by_users_a = (
    final_data[final_data['exp_group'] == 0]
    .groupby('session_start', as_index=False)
    .agg({'order_number': pd.Series.nunique})
)
orders_by_users_a.columns = ['visitor_id', 'orders']

orders_by_users_b = (
    final_data[final_data['exp_group'] == 1]
    .groupby('session_start', as_index=False)
    .agg({'order_number': pd.Series.nunique})
)
orders_by_users_b.columns = ['visitor_id', 'orders']

#Объявим переменные sample_a и sample_b,
#в которых пользователям из разных групп будет соответствовать количество заказов.
#Тем, кто ничего не заказал, будут соответствовать нули.
#Это нужно, чтобы подготовить выборки к проверке критерием Манна-Уитни.

sample_a = pd.concat(
    [orders_by_users_a['orders'],pd.Series(0, index=np.arange(data['visitors_per_date_a'].sum() - len(orders_by_users_a['orders'])), name='orders')],
    axis=0
)

sample_b = pd.concat(
    [orders_by_users_b['orders'],pd.Series(0, index=np.arange(data['visitors_per_date_b'].sum() - len(orders_by_users_b['orders'])), name='orders')]
    ,axis=0
) 

#Чтобы получить конверсию в группе, поделим сумму заказов на число пользователей
print("{0:.3f}".format(st.mannwhitneyu(sample_a, sample_b)[1]))

#Выведем относительный прирост конверсии группы B
print("{0:.3f}".format(sample_b.mean() / sample_a.mean() - 1))
```

Первое число — **p-value = 0.786** значительно больше уровня значимости 0.05. Значит, нулевую гипотезу о том, что статистически значимых различий в конверсии между группами нет, мы не можем отвергнуть. Прирост конверсии группы B относительно группы А составляет **6.2%**. 

Тест работает корректно, но при этом, по полученным данным, мы не добрали конверсию в 10%, как ожидалось. Она меньше - 6.2%.

### Расчет статистической значимости различия в выручке между группами

Сформулируем гипотезы.

`Нулевая`: различий в выручке между группами нет.

`Альтернативная`: различия в выручке между группами есть.

```python
revenue_final_data = final_data.query('revenue.notna()')

print('{0:.10f}'.format(st.mannwhitneyu(revenue_final_data[revenue_final_data['exp_group'] == 0]['revenue'], revenue_final_data[revenue_final_data['exp_group'] == 1]['revenue'])[1]))
print('{0:.3f}'.format(revenue_final_data[revenue_final_data['exp_group'] == 1]['revenue'].sum()/revenue_final_data[revenue_final_data['exp_group'] == 0]['revenue'].sum()-1)) 
```

P-value значительно ниже 0.05, что означает, что есть все причины отвергать нулевую гипотезу и считать, что в выручке групп А и В различий нет. Они есть. И второй показатель - 12.6% показывает, на сколько выручка группы В больше выручки группы А. В данном случае, мы убедились, что выручка больше 10%, как и ожидалось.
